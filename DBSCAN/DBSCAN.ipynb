```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# numpy – for numerical operations.
# matplotlib.pyplot – for plotting the clusters.
# make_moons – generates a synthetic dataset shaped like two interlocking moons (good for non-linear clustering).
# DBSCAN – the clustering algorithm from sklearn.
# StandardScaler – standardizes features by removing the mean and scaling to unit variance.
```

```python
#creates synthetic dataset
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)
```

```python
# X: the feature matrix (2D points).
# y: the true labels (not used in DBSCAN, but useful for evaluation).
# noise=0.05: adds some randomness to make the data more realistic.
```

```python
#feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Why scale? DBSCAN is distance-based. Without scaling, features with large values dominate.
# StandardScaler makes sure all features are on the same scale (mean = 0, std = 1).
```

```python
#apply dbscan
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)
```

```python
# eps=0.3: the radius for neighborhood. A point is a core point if it has at least min_samples within eps.
# min_samples=5: the minimum number of neighbors a point must have to be considered a core point.
# fit_predict: fits the model and returns the cluster labels for each point.
# Cluster labels:
# 0, 1, etc. → Cluster indices.
# -1 → Noise (outlier points)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='Paired', s=60)
plt.title("DBSCAN Clustering on make_moons Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# Colors represent different clusters.
# Points labeled -1 (noise) usually appear in gray/black.
```
