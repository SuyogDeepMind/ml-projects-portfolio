```python
## Step 4:-Checking the outliers
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1

for column in data.drop('Outcome',axis=1):
    if plotnumber<=9:
        ax=plt.subplot(3,3,plotnumber)
        sns.boxplot(data[column])
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Count',fontsize=20)
    plotnumber+=1
plt.tight_layout()
plt.show()
```

```python
#IQR
Q1=data['Pregnancies'].quantile(0.25)
print('Lower quantile',Q1)
Q3=data['Pregnancies'].quantile(0.75)
print('Upper quantile',Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
print('Lower limit: ',lower_limit)
upper_limit=Q3+1.5*IQR
print('Upper limit: ',upper_limit)
```

```python
## Step 5:-Scaling the data

from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
dl=['Pregnancies','Outcome']
data1=sc.fit_transform(data.drop(dl,axis=1))
```

```python
data2=pd.DataFrame(data1,columns=['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
con_data=data[['Pregnancies','Outcome']]
```

```python
con_data
```

```python
final_df=pd.concat([data2,con_data],axis=1)
```

```python
X=final_df.iloc[:,:-1]
y=final_df.Outcome
```

```python
y
```

### ðŸ”€ Insight: Data Splitting
Splitting the dataset into training and testing sets helps evaluate model performance on unseen data.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=45)
```

```python
X_test
```

### ðŸ§  Insight: Logistic Regression Model
This step initializes and trains the Logistic Regression model, ideal for binary classification problems.

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train,y_train)  ## training
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
y_pred_prob=clf.predict_proba(X_test)
```
